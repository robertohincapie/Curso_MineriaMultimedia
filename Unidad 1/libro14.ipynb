```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import torch

m=torch.tensor(5.0, requires_grad=True)
b=torch.tensor(2.0, requires_grad=True)
error=(3.1-b)**2+(3.4-m-b)**2+(3.7-2*m-b)**2
print(m, b, error)

grad=np.array([-21.6+10*5+6*2, -20.4+6*5+6*2])
print('Gradiente: ', grad)
```

```python
error.backward()
print(m.grad, b.grad)
```

```python
m=torch.tensor(1.0, requires_grad=True)
b=torch.tensor(1.0, requires_grad=True)
error=((m**2+b**2+torch.cos(3*m+b))**2-torch.sqrt(m**2+3*b**2))/torch.exp(m**2+b)
print(error)
error.backward()
print(m.grad, b.grad)
```

```python
#Proceso de la regresión lineal:
alfa=0.001
m=torch.tensor(5.0, requires_grad=True)
b=torch.tensor(2.0, requires_grad=True)
for i in range(5000):
  error=(3.1-b)**2+(3.4-m-b)**2+(3.7-2*m-b)**2
  m.grad=None
  b.grad=None
  error.backward()
  with torch.no_grad():
    m-=alfa*m.grad
    b-=alfa*b.grad
  #print(m,b,error)

print(m, b, error)
```

```python
x=torch.tensor(3.0, requires_grad=True)
x.grad=None
y=3+4*x-5*x**2
y.backward()
```
