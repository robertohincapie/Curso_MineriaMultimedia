# Introducción a Tensores en PyTorch

Operaciones aritméticas básicas.

```python
import torch

a=torch.tensor([1.,2.,3.])
b=torch.tensor([4.,5.,6.])
print(a)
print(b)
```

```python
# Suma
a+b
```

```python
# Resta
a-b
```

```python
# Multiplicación elemento a elemento
a*b
```

```python
# División
a/b
```

```python
# Escalares
2*a+1
```

```python
print(a.sum())
print(a.mean())
```

```python
A=torch.tensor([[1.,2.],[3.,4.]])
B=torch.tensor([[5.,6.],[7.,8.]])
A@B
```

```python
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
x=torch.tensor(3.5, requires_grad=True)
y=torch.tensor(1.5, requires_grad=True)
z=x*+2+y**2-4*x*y*torch.cos(3*x+2*y)
print('z=',z)
z.backward()
print('dz/dx=',x.grad, ', dz/dy=',y.grad)
```
