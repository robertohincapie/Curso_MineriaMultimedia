```python
def clean(x):
  return x.split(' ')[0][1:]
def visualizar(d):
  G=nx.DiGraph()
  G.add_node(str(d), label=str(d), shape='square', color='#fcf3cf')
  revisar=[(str(d), d.grad_fn)]
  labels={}
  finales=[]
  while(len(revisar)>0):
    parent, nodo=revisar.pop()
    #print(nodo)
    G.add_node(str(nodo), label=clean(str(nodo)))
    if parent is not None:
      G.add_edge(str(parent), str(nodo), shape='dot', color='#aed6f1')
    if(hasattr(nodo, 'next_functions')):
      if(len(nodo.next_functions)>0):
        #print('Nodo, next functions:',nodo.next_functions)
        for ni,_ in nodo.next_functions:
          #print(ni)
          revisar.append((str(nodo), ni))
      else:
        label=str(nodo.variable)
        name=str(nodo)+label
        G.add_node(name, label=label, shape='square', color='#d5f5e3')
        G.add_edge(str(nodo), name)
        finales.append((nodo,name))

  nt=Network(width='100%', height='600px', bgcolor='#dddddd', notebook=False, directed =True, cdn_resources='remote')#, filter_menu=True)
  nt.from_nx(G.reverse())
  nt.options.edges = {"arrows": {"to": {"enabled": True, "scaleFactor": 1}}}
  nt.write_html("temp.html")
  with open("temp.html", "r") as file:
    html_string = file.read()
  return html_string
```

```python
x=torch.tensor(3.0, requires_grad=True)
y=torch.tensor(1.0, requires_grad=True)
z=3*x*torch.exp(x**2+y**2)*torch.cos(x/y)+torch.tanh(x**2+y)
html_string=visualizar(z)
HTML(html_string)
```

```python
x=torch.tensor(2.0, requires_grad=True)
y=torch.tensor(-1.0, requires_grad=True)

z=x**3+3*x*y+y**2+2*torch.cos(x**2+torch.tan(y/4))/(x**2+y**2)
html_string=visualizar(z)
HTML(html_string)
```
