```python
from torch import nn as nn
import torch
#Con Kernels de 3x3
m = nn.Conv2d(40, 100, 5, stride=1, padding=2) #40, numero de capas de entrada. 100 es el numero de capas de salida.
                                              #    5 representa un filtro de 5x5
input = torch.randn(20, 40, 50, 100)          #20 número de imagenes, batch 20 imagenes.
                                              #40 número de capas de la imagen.
                                              #50, 100 es la dimension de la imagen.
output = m(input)
print(input.shape)
print(output.shape)
```

```python
m.weight.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
for i in range(16):
  plt.subplot(4,4,i+1)
  plt.imshow(m.weight[0,i,:,:].detach().numpy())
```

```python
m = nn.MaxPool2d(2, stride=2, padding=0)
output2 = m(output)
print(output.shape)
print(output2.shape)
```
